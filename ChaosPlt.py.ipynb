```python
# plotting the motion of a 2D spring with natural length of l and stretched length of d
# this system has Chaotic behavior which is sensitive to the initial conditions 

import matplotlib.pyplot as plt
import numpy as np

# springs constant
k = 80

# mass of object between springs
m = 10

# frequency 
w=(k/m)**0.5

# natural length
l = 1

# stretched length
d = 1.5

# initial position in x axis
x = 0.1

# initial position in y axis
y =0.1

# initial velocity in x axis
vx=0

# initial velocity in y axis
vy=0

# time
t=0
dt=0.01

xRange =[]
yRange = []

while t<100:
    t+=dt
    x+=vx*dt
    y+=vy*dt
    d1=(x**2+(d-y)**2)**0.5
    d2=(x**2+(d+y)**2)**0.5
    ax=-(w**2)*(2*x - (x*l)*(1/d1 + 1/d2))
    ay= -(w**2)*(2*y - l*((d+y)/d2 + (y-d)/d1))
    vx+=ax*dt
    vy+=ay*dt
    xRange.append(x)
    yRange.append(y)
plt.plot(xRange,yRange)
```

```python
# forced oscillation with large domain
# here I choose x insted of theta

import matplotlib.pyplot as plt
import numpy as np

c = 0.05
F = 0.4
y =0.01
w = 0.7
x = 0.001
t = 0
dt = 0.01 #np.pi/50
z = 0

xRange = []
yRange = []

while t<100: #np.pi/0.005
    t+=dt
    z+=w*dt
    x+=y*dt
    ydot=-c*y -np.sin(x) + F*np.cos(z)
    y+=ydot*dt
    xRange.append(z)
    yRange.append(y)

plt.plot(xRange , yRange)
```
