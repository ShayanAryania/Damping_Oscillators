```python
# plotting the damped harmonic oscillator 

import numpy as np
import matplotlib.pyplot as plt

# first domain
A = 1

# damping ratio
betha = 0.4

# phase difference
alpha = np.pi/2

# natural frequency
w0=np.pi/0.5

# damping frequency
w1 = (w0**2 - betha**2)**0.5

#time steps
t = np.arange(0,10,0.001)

xRange=[]
yRange=[]

for i in range(len(t)):
    x = A*np.exp(-betha*t[i])*(np.sin(w1*t[i] - alpha))
    xRange.append(i)
    yRange.append(x)
    
plt.plot(xRange , yRange)
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show() 
```

```python
# same program with function

import numpy as np
import matplotlib.pyplot as plt

def osi(t):
    A = 1
    betha = 0.4
    alpha = np.pi/2
    w0=np.pi/0.5
    w1 = (w0**2 - betha**2)**0.5
    x = A*np.exp(-betha*t)*(np.sin(w1*t- alpha))
    return x

t = np.arange(0,10,0.001)
plt.plot(t , osi(t))
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show() 
```

```python
# phase space for the damped oscillator

import numpy as np
import matplotlib.pyplot as plt

A = 1

betha = 0.4

alpha = np.pi/2

w0= np.pi/0.9

w1 = (w0**2 - betha**2)**0.5

t = np.arange(0,10,0.01)

xRange=[]
yRange=[]

for i in range(len(t)):
    x = A*np.exp(-betha*t[i])*(np.sin(w1*t[i] - alpha))
    xdot = (-betha)*A*np.exp(-betha*t[i])*(np.sin(w1*t[i] - alpha)) + A*w1*np.exp(-betha*t[i])*np.cos(w1*t[i] - alpha)
    xRange.append(x)
    yRange.append(xdot)
    
plt.plot(xRange , yRange)
plt.xlim(-2.5, 2.5)
plt.ylim(-2.5, 3)
plt.xlabel('x (m)')
plt.ylabel('xdot (m/s)')
plt.draw()
```

```python
#SHM phase plane
import matplotlib.pyplot as plt
import numpy as np

w = 1
t = np.arange(0,10,0.001)

xRange = []
yRange = []

for i in range(len(t)):
    x = np.cos(w*t[i])
    xdot = -w*(np.sin(w*t[i]))
    xRange.append(x)
    yRange.append(xdot)
    
plt.plot(xRange,yRange)
plt.xlabel('x (m)')
plt.ylabel('xdot (m/s)')
```

```python
# forced oscillation

import numpy as np
import matplotlib.pyplot as plt

# initial force
F0 = 2

# first domain
A = 2

# damping ratio
betha = 0.15

# phase difference
alpha = np.pi

# natural frequency
w0=np.pi

# damping frequency
w1 = (w0**2 - betha**2)**0.5

# force frequency
w = 1/2*w1

# time
t = np.arange(0,50,0.001)

xRange=[]
yRange=[]

for i in range(len(t)):
    # general solution of eq
    xc = A*np.exp(-betha*t[i])*(np.cos(w1*t[i] - alpha))
    felan = w0**2 - w**2 + 4*(betha**2)*(w**2)
    # particular solution of eq
    xp = F0*np.cos(w*t[i])/(felan**0.5)
    xRange.append(t[i])
    yRange.append(xc+xp)

x1Range=[]
y1Range=[]

for i in range(len(t)):
    felan = w0**2 - w**2 + 4*(betha**2)*(w**2)
    xp = F0*np.cos(w*t[i])/(felan**0.5)
    x1Range.append(t[i])
    y1Range.append(xp)
    
plt.plot(xRange , yRange)
plt.plot(x1Range , y1Range,'--')
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show() 
```

```python
# 2D oscillator
# Lissajous curves

import numpy as np
import matplotlib.pyplot as plt

xRange = []
yRange = []

# domain of x axis spring
A = 1

# domain of y axis spring
B = 2

# frequency of x axis spring
wx= np.pi/0.25

# frequency of y axis spring
wy= (3/4)*wx

# time
t = np.arange(0,10,0.001)

for i in range(len(t)):
    x = A*np.cos(wx*t[i]+np.pi/2)
    y = B*np.cos(wy*t[i]+np.pi/2)
    xRange.append(x)
    yRange.append(y)
    
    
plt.plot(xRange,yRange)
plt.xlabel('x(m)')
plt.ylabel('y(m)')
plt.show()
```

```python
# plotting the damped harmonic oscillator 
# without solving eq / plot by the main eq: a + 2(betha)v + w^2(x) = 0 

import numpy as np
import matplotlib.pyplot as plt

# first velocity
v0x = 5

# first position
x = 0.1

# spring constant
k = 5

# mass of object
m = 0.25

# frequency
w=(k/m)**0.5

# damping ratio
betha = 0.25

# time
t =0
deltaT=0.001

xRange = []
yRange = []

while t<20:
    t+=deltaT
    x+=v0x*deltaT
    ax = -2*betha*v0x - (w**2)*(x)
    v0x+=ax*deltaT
    xRange.append(t)
    yRange.append(x)
    
plt.plot(xRange , yRange)
plt.xlabel('time (s)')
plt.ylabel('x (m)')
plt.show()
```
